```python
# ================================================================================
# @file    dimensionality_reduction.m
# @brief   Dimensionality reduction and classification of structure data
# @author  [Yuqing Li]
# @date    2026-03-03
# @version 1.0
# @license MIT (See LICENSE file for details)
# @contact [202321250018@mail.bnu.edu.cn]
# @note    The 91-region structural data were embedded into 2d
# low-dimensional spaces using t-SNE, CMDS, or UMAP, then classified into
# different groups by k-means or GMM.


import pandas as pd
import numpy as np
from scipy.io import loadmat, savemat
import umap
import os

# data loading
rvariance = loadmat('DDist_UMAP.mat')
variance = rvariance['DDist_UMAP']

print(f"input shape: {variance.shape}")
print(f"dtype: {type(variance)}")

# cell-wise processing
all_results = []

for idx in range(variance.shape[0]):
    
    variance_matrix = np.array(variance[idx])[0]
    print(f"\nprocessing {idx+1}/{variance.shape[0]}, shape: {variance_matrix.shape}")
    
    nsize = variance_matrix.shape[0]
    total_Dist = np.zeros((nsize, nsize))
    
    # UMAP reduction with 500 iterations
    for i in range(500):
        reducer = umap.UMAP(n_components=2, random_state=42+i)
        embedding = reducer.fit_transform(variance_matrix)
        
        from scipy.spatial.distance import pdist, squareform
        dist_vec = pdist(embedding, metric='euclidean')
        Dist = squareform(dist_vec)
        
        total_Dist += Dist
        
        if (i+1) % 500 == 0:
            print(f"  iter {i+1}/100 done")
    
    total_Dist = total_Dist / 500
    all_results.append(total_Dist)

# output configuration
output_folder = "/"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

output_mat_path = os.path.join(output_folder, 'All_UMAP.mat')

# matlab cell array conversion
output_cell = np.empty((len(all_results),), dtype=object)
for i, arr in enumerate(all_results):
    output_cell[i] = arr

savemat(output_mat_path, {'All_UMAP': output_cell})

print(f"\nsaved: {output_mat_path}")
print("done")
```
